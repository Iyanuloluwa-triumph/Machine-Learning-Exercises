```python
df[df["Cholesterol"] >(df.Cholesterol.mean() + 3*df.Cholesterol.std())]
```

```python
df1 = df[df["Cholesterol"] <(df.Cholesterol.mean() + 3*df.Cholesterol.std())]
df1.head()
```

```python
df[df["RestingBP"] >(df.RestingBP.mean() + 3*df.RestingBP.std())]
```

```python
df2 = df1[df1["RestingBP"] <(df.RestingBP.mean() + 3*df1.RestingBP.std())].copy()
df2
```

```python
df[df["MaxHR"] >(df.MaxHR.mean() + 3*df.MaxHR.std())]
```

```python
df[df["Oldpeak"] <=(df.Oldpeak.mean() - 3*df.Oldpeak.std())]
```

```python
df3 = df2[df2["Oldpeak"] >=(df.Oldpeak.mean() - 3*df.Oldpeak.std())].copy()
df3
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df3["Sex"] = le.fit_transform(df3["Sex"])
df3.ExerciseAngina = le.fit_transform(df3["ExerciseAngina"])
df3
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_features = ["ChestPainType", "RestingECG", "ST_Slope"]
transformed = ColumnTransformer(
     [(
        "cat_features", OneHotEncoder(drop="first"), categorical_features
    )], remainder="passthrough"
)
```

```python
final = transformed.fit_transform(df3)
final
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
data = scaler.fit_transform(final[:, :-1])
data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
data
```

```python
from sklearn.model_selection import cross_val_score
svm_score = np.mean(cross_val_score(SVC(),  data, final[:, -1], cv=5))
lg_score = np.mean(cross_val_score(LogisticRegression(),  data, final[:, -1], cv=5))
forest_score = np.mean(cross_val_score(RandomForestClassifier(n_estimators=150),  data, final[:, -1], cv=5))
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(2)
pca_data = pc.fit_transform(data)
```

```python
svm_score = np.mean(cross_val_score(SVC(), pca_data, final[:, -1], cv=5))
svm_score
```
