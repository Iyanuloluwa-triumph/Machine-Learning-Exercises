```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# OneHotEncode ChestPainType
encoder = OneHotEncoder(sparse_output=False, drop="first")
X_transformed = encoder.fit_transform(df2[["ChestPainType", "RestingECG"]])

# Get the feature names directly from the encoder
feature_names = encoder.get_feature_names_out(["ChestPainType", "RestingECG"])

# Make DataFrame with correct column names
df_encoded = pd.DataFrame(X_transformed, columns=feature_names)
```

```python
# Get the feature names directly from the encoder
feature_names = encoder.get_feature_names_out(["ChestPainType", "RestingECG"])
```

```python
# Make DataFrame with correct column names
df_encoded = pd.DataFrame(X_transformed, columns=feature_names)
df_encoded[300:500]
```

```python
merged_df = pd.concat([df2.reset_index(drop=True), df_encoded.reset_index(drop=True)], axis="columns")
merged_df
```

```python
merged_df.drop(merged_df[["ChestPainType", "RestingECG"]], axis="columns", inplace=True)
merged_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_transformed = scaler.fit_transform(merged_df.iloc[:, :-1])
X_transformed
```

```python
y = merged_df.iloc[:, -1]
y
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
svm_score = np.mean(cross_val_score(SVC(), X_transformed, y, cv=5))
svm_score
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
svm_bagged = BaggingClassifier(
    estimator= SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score = True,
    random_state = 42
    
)
svm_bagged.fit(X_transformed, y)
```

```python
svm_bagged.oob_score_
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_bagged = BaggingClassifier(
    estimator= DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score = True,
    random_state = 42
    
)
tree_bagged.fit(X_transformed, y)
```

```python
tree_bagged.oob_score_
```

```python
tree_cv = np.mean(cross_val_score(DecisionTreeClassifier(), X_transformed, y, cv=5))
tree_cv
```
